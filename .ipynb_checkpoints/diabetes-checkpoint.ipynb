```python
# Deteksi outliers pakai IQR method
def detect_outliers(df, features):
    outlier_indices = []
    for feature in features:
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        feature_outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)].index
        outlier_indices.extend(feature_outliers)
    
    return list(set(outlier_indices))

# Features untuk outlier detection (exclude ID columns)
outlier_features = [col for col in df.select_dtypes(include=[np.number]).columns 
                   if col not in ['ID', 'No_Pation']]

outliers = detect_outliers(df, outlier_features)
print(f"Outliers detected: {len(outliers)} rows ({len(outliers)/len(df)*100:.1f}%)")

# Visualize outliers untuk key features
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
key_features = ['HbA1c', 'BMI', 'Chol', 'TG']

for i, feature in enumerate(key_features):
    row, col = i//2, i%2
    sns.boxplot(data=df, y=feature, ax=axes[row, col])
    axes[row, col].set_title(f'Outliers in {feature}')

plt.tight_layout()
plt.show()
```

**Analysis Outlier**

- HbA1c â†’ ada outlier tinggi (> 14) dan rendah (< 2).
- BMI â†’ ada beberapa pasien dengan BMI > 45 (jarang tapi masih mungkin secara medis).
- Chol (Kolesterol) â†’ banyak outlier tinggi, bahkan > 9.
- TG (Trigliserida) â†’ outlier ekstrim sampai 14 (normal < 2). <br>

Statistik deteksi: 321 rows (32,1%) outlier â†’ cukup banyak.

```python
# Outliers handling

# Fungsi capping (winsorization)
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # batasi nilai ke ambang bawah/atas
    df[column] = np.where(df[column] < lower_bound, lower_bound,
                          np.where(df[column] > upper_bound, upper_bound, df[column]))
    return df

# capping untuk kolom numeric penting
features_to_cap = ['HbA1c', 'BMI', 'Chol', 'TG']
for col in features_to_cap:
    df = cap_outliers(df, col)

print("Outliers dicapping ke ambang bawah/atas")
```

```python
# Visualisasi after capping
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
for i, feature in enumerate(features_to_cap):
    row, col = i//2, i%2
    sns.boxplot(data=df, y=feature, ax=axes[row, col], color="skyblue")
    axes[row, col].set_title(f'{feature} after capping')

plt.tight_layout()
plt.show()
```

```python
# feature engineering
print("Creating new features...")

# 1. Cholesterol Risk Ratio
df['Chol_HDL_Ratio'] = df['Chol'] / df['HDL']
df['LDL_HDL_Ratio'] = df['LDL'] / df['HDL']

# 2. Age Groups
df['Age_Group'] = pd.cut(df['AGE'], bins=[0, 30, 45, 60, 100], 
                        labels=['Young', 'Middle_Age', 'Senior', 'Elderly'])

# 3. BMI Categories
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI_Category'] = df['BMI'].apply(bmi_category)

# 4. Diabetes Risk Score (custom medical composite score)
df['Diabetes_Risk_Score'] = (df['HbA1c'] * 0.4 + 
                             df['BMI'] * 0.2 + 
                             df['TG'] * 0.1 + 
                             df['AGE'] * 0.1 + 
                             (1/df['HDL']) * 0.2)

# 5. Metabolic Syndrome Indicators
df['High_TG'] = (df['TG'] >= 150).astype(int)  # High triglycerides
df['Low_HDL'] = ((df['Gender'] == 'M') & (df['HDL'] < 40) | 
                 (df['Gender'] == 'F') & (df['HDL'] < 50)).astype(int)
df['High_BP_Proxy'] = (df['AGE'] > 45).astype(int)  # Age as proxy for hypertension risk

print("New features created:")
new_features = ['Chol_HDL_Ratio', 'LDL_HDL_Ratio', 'Age_Group', 'BMI_Category', 
                'Diabetes_Risk_Score', 'High_TG', 'Low_HDL', 'High_BP_Proxy']
for feature in new_features:
    print(f"   - {feature}")
```

- Cholesterol Ratios: Rasio kolesterol lebih informatif daripada nilai absolut
- Age Groups: Kategorisasi usia membantu model memahami pola non-linear
- BMI Categories: WHO standard categories
- Diabetes Risk Score: Composite score berdasarkan medical knowledge
- Metabolic Syndrome: Binary indicators untuk kondisi terkait diabetes

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode categorical variables
le_gender = LabelEncoder()
df['Gender_Encoded'] = le_gender.fit_transform(df['Gender'])

# One-hot encoding for multi-category features
df = pd.get_dummies(df, columns=['Age_Group', 'BMI_Category'], prefix=['Age', 'BMI'])

print("Data shape after encoding:", df.shape)
print("New columns added:")
print([col for col in df.columns if col.startswith(('Age_', 'BMI_'))])
```

### Model development

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Prepare features and target
# Drop non-predictive columns
columns_to_drop = ['ID', 'No_Pation', 'CLASS', 'Gender', 'Age_Group', 'BMI_Category']
X = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
y = df['CLASS']

print("Features selected:", X.shape[1])
print("Feature names:")
print(X.columns.tolist())

# Split data with stratification (maintain class distribution)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    stratify=y, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f" Training set: {X_train.shape[0]} samples")
print(f" Test set: {X_test.shape[0]} samples")
print(f" Class distribution in training:")
print(y_train.value_counts(normalize=True))
```

```python
# baseline model comparison

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

# inisiasi models
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM': SVC(random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42)
}

# Cross-validation results
cv_results = {}
cv_folds = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print(" Cross-Validation Results (5-fold):")
print("-" * 50)

for name, model in models.items():
    # Use scaled data for models that need it
    if name in ['Logistic Regression', 'SVM', 'K-Nearest Neighbors']:
        scores = cross_val_score(model, X_train_scaled, y_train, cv=cv_folds, scoring='accuracy')
    else:
        scores = cross_val_score(model, X_train, y_train, cv=cv_folds, scoring='accuracy')
    
    cv_results[name] = {
        'mean': scores.mean(),
        'std': scores.std()
    }
    
    print(f"{name:20}: {scores.mean():.3f} (+/- {scores.std() * 2:.3f})")

# Sort results
sorted_results = sorted(cv_results.items(), key=lambda x: x[1]['mean'], reverse=True)
print(f"\n Best performing model: {sorted_results[0][0]} with {sorted_results[0][1]['mean']:.3f} accuracy")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, f1_score

# Select best model for hyperparameter tuning
best_model_name = sorted_results[0][0]
print(f" Optimizing: {best_model_name}")

# Define parameter grids
param_grids = {
    'Random Forest': {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'Gradient Boosting': {
        'n_estimators': [100, 200],
        'learning_rate': [0.05, 0.1, 0.2],
        'max_depth': [3, 5, 7],
        'subsample': [0.8, 1.0]
    },
    'Logistic Regression': {
        'C': [0.1, 1, 10, 100],
        'penalty': ['l1', 'l2'],
        'solver': ['liblinear', 'saga']
    }
}

# Get appropriate model and parameters
if best_model_name in param_grids:
    best_base_model = models[best_model_name]
    param_grid = param_grids[best_model_name]
    
    # Use macro F1-score for multiclass evaluation
    scorer = make_scorer(f1_score, average='macro')
    
    # Grid search with cross-validation
    grid_search = GridSearchCV(
        best_base_model, 
        param_grid, 
        cv=5, 
        scoring=scorer,
        n_jobs=-1,
        verbose=1
    )
    
    # Fit with appropriate data
    if best_model_name in ['Logistic Regression', 'SVM', 'K-Nearest Neighbors']:
        grid_search.fit(X_train_scaled, y_train)
        X_train_final = X_train_scaled
        X_test_final = X_test_scaled
    else:
        grid_search.fit(X_train, y_train)
        X_train_final = X_train
        X_test_final = X_test
    
    # Best model
    best_model = grid_search.best_estimator_
    
    print(f"Best parameters: {grid_search.best_params_}")
    print(f"Best CV score: {grid_search.best_score_:.3f}")
else:
    # If no param grid defined, use the base model
    best_model = models[best_model_name]
    if best_model_name in ['Logistic Regression', 'SVM', 'K-Nearest Neighbors']:
        best_model.fit(X_train_scaled, y_train)
        X_train_final = X_train_scaled
        X_test_final = X_test_scaled
    else:
        best_model.fit(X_train, y_train)
        X_train_final = X_train
        X_test_final = X_test
```

### Model Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

# Make predictions
y_pred = best_model.predict(X_test_final)
y_pred_proba = best_model.predict_proba(X_test_final) if hasattr(best_model, 'predict_proba') else None

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f" Test Accuracy: {accuracy:.3f}")

# Detailed classification report
print("\n Detailed Classification Report:")
print("="*60)
report = classification_report(y_test, y_pred, output_dict=True)
print(classification_report(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Normal', 'Pre-diabetes', 'Diabetes'],
            yticklabels=['Normal', 'Pre-diabetes', 'Diabetes'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Calculate per-class metrics
classes = ['N', 'P', 'Y']  # Adjust based on your class labels
class_names = ['Normal', 'Pre-diabetes', 'Diabetes']

print("\n Per-Class Performance:")
print("-"*50)
for i, (class_code, class_name) in enumerate(zip(classes, class_names)):
    if class_code in report:
        precision = report[class_code]['precision']
        recall = report[class_code]['recall']
        f1 = report[class_code]['f1-score']
        support = report[class_code]['support']
        
        print(f"{class_name:12}: Precision={precision:.3f}, Recall={recall:.3f}, F1={f1:.3f}, n={support}")
```

```python
# Feature importance (for tree-based models)
if hasattr(best_model, 'feature_importances_'):
    # Get feature importance
    feature_importance = pd.DataFrame({
        'feature': X.columns,
        'importance': best_model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    # Plot top 15 features
    plt.figure(figsize=(10, 8))
    top_features = feature_importance.head(15)
    sns.barplot(data=top_features, x='importance', y='feature', palette='viridis')
    plt.title('Top 15 Most Important Features')
    plt.xlabel('Feature Importance')
    plt.tight_layout()
    plt.show()
    
    print(" Top 10 Most Important Features:")
    print("="*50)
    for idx, row in feature_importance.head(10).iterrows():
        print(f"{row['feature']:20}: {row['importance']:.4f}")
        
elif hasattr(best_model, 'coef_'):
    # For linear models, show coefficients
    feature_coef = pd.DataFrame({
        'feature': X.columns,
        'coefficient': np.abs(best_model.coef_[0])  # Take absolute value
    }).sort_values('coefficient', ascending=False)
    
    plt.figure(figsize=(10, 8))
    top_features = feature_coef.head(15)
    sns.barplot(data=top_features, x='coefficient', y='feature', palette='viridis')
    plt.title('Top 15 Features by Coefficient Magnitude')
    plt.xlabel('|Coefficient|')
    plt.tight_layout()
    plt.show()
```

```python
# Analyze prediction errors
error_df = pd.DataFrame({
    'True_Class': y_test,
    'Predicted_Class': y_pred,
    'Correct': y_test == y_pred
})

# Add original features for error analysis
error_df = pd.concat([error_df, X_test.reset_index(drop=True)], axis=1)

print("âŒ Error Analysis:")
print("="*40)

# Most common errors
error_summary = error_df[~error_df['Correct']].groupby(['True_Class', 'Predicted_Class']).size()
print("\nðŸš¨ Most Common Prediction Errors:")
for (true_class, pred_class), count in error_summary.items():
    total_true = (y_test == true_class).sum()
    error_rate = count / total_true * 100
    print(f"{true_class} â†’ {pred_class}: {count} cases ({error_rate:.1f}% of all {true_class})")

# Analyze characteristics of misclassified cases
print("\nðŸ” Characteristics of Misclassified Cases:")
misclassified = error_df[~error_df['Correct']]
correctly_classified = error_df[error_df['Correct']]

key_features_analysis = ['HbA1c', 'BMI', 'AGE', 'Chol', 'TG', 'HDL']
for feature in key_features_analysis:
    if feature in misclassified.columns:
        misc_mean = misclassified[feature].mean()
        corr_mean = correctly_classified[feature].mean()
        print(f"{feature:10}: Misclassified avg={misc_mean:.2f}, Correct avg={corr_mean:.2f}")
```
